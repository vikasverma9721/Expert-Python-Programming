***Generator Functions :***

***Create a generator function named product_details() that yields three specific pieces of information about a laptop in sequential order.***

```python
def product_details():
    yield "Lenovo"
    yield 16
    yield 45000.

laptop = product_details()
```

```python
print(next(laptop))
```

***Ques-2. Generator uses strings instead of numbers, but the behavior of the for loop remains the same.***

```python
def product_details():
  yield "Brand: Audi"
  yield "Model: 2025"
  yield "Mileage: 35Km/h"

laptop_info = product_details()

for info in laptop_info:
  print(info)
```

***Ques-4. Generator calculates the powers of 2 (2, 4, 8, 16...) for a given count n.***

```python
def power_of_two(n):
    value = 1
    for x in range(n):
      yield value
      value *= 2

gen_pow = power_of_two(5)

print(list(gen_pow))
```

***Ques-5. Generator performs multiplication based on the value received through the .send() method.***

```python
def multiplier(n):
  while True:
    factor = (yield n)
    if factor is not None:
      n *= factor
    else:
      n *= 2

m = multiplier(5)
print(next(m))
print(m.send(6))
print(next(m))
print(next(m))
```
