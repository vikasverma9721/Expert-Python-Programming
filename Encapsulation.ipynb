***Ques_01. Create a Laptop class with a private attribute __price. Implement a method set_price(new_price) to update the price and get_price() to retrieve the price.***


```python
class Laptop:
  def __init__(self, price):
    self.__price = price

  def set_price(self, new_price):
    if new_price > 0:
      self.__price = new_price
    else:
      print("Invalid price")

  def get_price(self):
    return self.__price

laptop1 = Laptop(60000)
print("Current price",laptop1.get_price())

laptop1.set_price(70000)
print("New price",laptop1.get_price())
```

***Ques_02. Create a BankAccount class with a private attribute __balance. Implement methods to deposit, withdraw, and check balance while ensuring the balance cannot be modified directly.***


```python
class BankAccount:
  def __init__(self, balance):
    self.__balance = balance

  def get_balance(self):
    return self.__balance

  def deposit(self, amount):
    if amount > 0:
      self.__balance += amount
      print(f'Deposit {amount}. new balance {self.__balance}')
    else:
        print("Invalid amount")

  def withdraw(self, amount):
    if amount <= self.__balance:
      self.__balance -= amount
      print('withdraw successfull')
    else:
      print("Insuffient amount")
    return self.__balance

acc1 = BankAccount(20000)
acc1.deposit(500)
acc1.withdraw(250)
print('Current balance',acc1.get_balance())
```

***Ques_03. Create a Movie class with a private attribute __rating. Implement methods to set and get the rating, ensuring the rating is between 1 and 10.***


```python
class Movie:
  def __init__(self, rating):
    self.__rating = max(1, min(10, rating))

  def set_rating(self, new_rating):
    if 1<= new_rating <= 10:
      self.__rating = new_rating

  def get_rating(self):
    return self.__rating
movie = Movie(11)
print('Rating is',movie.get_rating())
```

***Ques_04 Create a Book class with private attributes __title and __author. Implement a constructor to set these values and provide methods to retrieve them.***

```python
class Book:
  def __init__(self, title, author):
    self.__title = title
    self.__author = author

  def get_title(self):
    return self.__title

  def get_author(self):
    return self.__author

book = Book('Mahabharat', 'Vyasa')
print('Book title is',book.get_title())
print('Book author is',book.get_author())
```

***Ques_05. Create a Car class with a private attribute __speed. Implement methods to increase speed, decrease speed, and get the current speed, ensuring speed does not go below 0.***

```python
class Car:
  def __init__(self, speed):
    self.__speed = max(0, speed)

  def increase_speed(self, speed_value):
    if speed_value > 0:
      self.__speed += speed_value

  def decrease_speed(self, speed_value):
    if 0 < speed_value <= self.__speed:
      self.__speed -= speed_value

  def get_speed(self):
    return self.__speed

car1 = Car(40)
print(car1.get_speed())
car1.increase_speed(20)
print('Speed after increase', car1.get_speed())
car1.decrease_speed(10)
print('Speed befor decrease', car1.get_speed())
```
