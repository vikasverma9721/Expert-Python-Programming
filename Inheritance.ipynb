```python
class Diet:
    def __init__(self):
        self.calories = 0

    def junk_calories(self, amount):
        """
        Adds calories to the total count (eating junk food).
        """
        self.calories += amount
        print(f"Ate junk food: +{amount} calories.")

    def exercise(self, amount):
        """
        Subtracts calories from the total count (burning calories).
        """
        self.calories -= amount
        # Prevent negative calories
        if self.calories < 0:
            self.calories = 0
        print(f"Exercised: -{amount} calories.")

    def check_health_status(self):
        """
        Checks if the current calorie count is Healthy or Unhealthy.
        Threshold: 30 calories.
        """
        print(f"Current Calorie Count: {self.calories}")

        if self.calories > 30:
            print("Status: Unhealthy (Calories are greater than 30)")
        else:
            print("Status: Healthy (Calories are less than or equal to 30)")

# --- Example Usage ---

# 1. Create the object
my_diet = Diet()

# 2. Add some junk food (Initial state)
# Adding 10 calories (Total: 10) -> Should be Healthy (<30)
my_diet.junk_calories(10)
my_diet.check_health_status()

print("-" * 20)

# 3. Eat more junk food
# Adding 40 calories (Total: 50) -> Should be Unhealthy (>30)
my_diet.junk_calories(40)
my_diet.check_health_status()

print("-" * 20)

# 4. Exercise to burn it off
# Burning 30 calories (Total: 20) -> Should be Healthy again (<30)
my_diet.exercise(30)
my_diet.check_health_status()
```

```python
class RacingCar:
    def __init__(self):
        # Initialize speed as 0 as requested
        self.speed = 0

    def accelerate(self, acceleration_rate, time):
        """
        Increases speed based on acceleration and time.
        Formula: Speed = Current Speed + (Acceleration * Time)
        """
        increased_speed = acceleration_rate * time
        self.speed += increased_speed
        print(f"Accelerating... (Rate: {acceleration_rate} m/s², Time: {time}s)")

    def brake(self, deceleration_rate, time):
        """
        Decreases speed based on deceleration (braking) and time.
        Formula: Speed = Current Speed - (Deceleration * Time)
        """
        decreased_speed = deceleration_rate * time
        self.speed -= decreased_speed

        # logical check to ensure speed doesn't go below 0
        if self.speed < 0:
            self.speed = 0

        print(f"Braking... (Rate: {deceleration_rate} m/s², Time: {time}s)")

    def show_speed(self):
        print(f"Current Speed: {self.speed} m/s")

# --- Example Usage ---

# 1. Create the object
my_car = RacingCar()
my_car.show_speed()

# 2. Accelerate: acceleration of 5 m/s² for 10 seconds
# Calculation: 0 + (5 * 10) = 50
my_car.accelerate(5, 10)
my_car.show_speed()

# 3. Brake: deceleration of 2 m/s² for 5 seconds
# Calculation: 50 - (2 * 5) = 40
my_car.brake(2, 5)
my_car.show_speed()
```

```python
class IMDb:
    def __init__(self, movie_name, initial_rating=0):
        self.movie = movie_name
        self.rating = initial_rating

    def new_rating(self, updated_rating):
        """
        Updates the rating of the movie with a new value.
        """
        # Optional: Validate that the rating is within a standard 1-10 range
        if 0 <= updated_rating <= 10:
            self.rating = updated_rating
            print(f"Rating updated for '{self.movie}' to {self.rating}/10")
        else:
            print("Error: Rating must be between 0 and 10.")

    def show_details(self):
        """
        Helper method to display the movie and its current rating.
        """
        print(f"Movie: {self.movie} | Current Rating: {self.rating}/10")

# --- Example Usage ---

# 1. Create the object (The movie is "Inception")
favorite_movie = IMDb("Inception", 8.5)
favorite_movie.show_details()

# 2. Update the rating using the new_rating method
favorite_movie.new_rating(9.0)

# 3. Verify the change
favorite_movie.show_details()
```
