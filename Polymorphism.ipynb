***Polymorphism***

```python
class Bird:
  def fly(self):
    print('Birds can fly')

class Penguin:
  def fly(self):
    print('Penguin cannot fly')

b = Bird()
p = Penguin()

b.fly()
p.fly()
```

```python
class Cat:
  def sound(self):
    return 'Meow'

class Dog:
  def sound(self):
    return 'Woof'

animals = [Cat(), Dog()]
for animal in animals:
  print(animal.sound())
```

```python
class Vehicle:
  def speed(self):
    print('The Vehicle moves at the standard speed')

class Car(Vehicle):
  def speed(self):
    print('The car moves at a speed of 120 km/h.')

vehicle = Vehicle()
vehicle.speed()

car = Car()
car.speed()
```

```python
class Box:
  def __init__(self, length, width, height):
    self.length = length
    self.width = width
    self.height = height

  def Valume(self):
    return self.length * self.width * self.height


  def __add__(self, other):
    if isinstance(other, Box):
      return self.Valume() + other.Valume()
    return NotImplemented

box1 = Box(4,5,6)
box2 = Box(2, 3, 6)

total_valume = box1 + box2
print('Total Valume', total_valume)
```
