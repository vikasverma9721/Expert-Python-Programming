***Create an iterator from a list and print its first two elements using next.***

```python
vikas = [1,2,3,4,5]
iterator = iter(vikas)
```

```python
print(next(iterator))
print(next(iterator))
```

***Convert a string into an iterator and print each character using next.***

```python
text = 'Hello'
iterator = iter(text)
```

```python
print(next(iterator))
```

***By using for loop.***

```python
numbers = [1,2,3,4,5,6]
iterator = iter(numbers)

for item in iterator:
  print(item)
```

***How to you use a for loop to iterate over an iterator.***

```python
letters = iter(["a","b","c"])
for  letter in letters:
  print(letter)
```

***How do you use iter() and next() to iterate through a range of numbers?***

```python
range_iter = iter(range(6))
print(next(range_iter))
print(next(range_iter))
```

```python
print(range(6))
```

***Manually iterate through a tuple using an iterator and a while loop.***

```python
fruits = ('apple','banana','cherry')
fruits_iter = iter(fruits)

while True:
   try:
      print(next(fruits_iter))
   except StopIteration:
       break
```
