***Decorator.***

```python
# with decorator

def decor(func):
  def wrapper():
    print("Not start")
    func()

    print("Start")
  return wrapper

@decor
def printer():
  print("Hello")
  print("2Hello")

printer()
```

```python
def decor(func):
  def inner():
    func()
    print("1Welcome")
  return inner

@decor
def printer():
  print("2Welcome")
  print("3Welcome")

printer()
```

```python
def Decorator_Function(func):  # Create decorator function
  def wrapper():  #
    print("Before the original function")
    func()  #original function
    print("After the original function")
  return wrapper

@Decorator_Function
def original_function():
  print("Inside original function")
  pass

original_function()  #call the function
```

## Logging Decorator.

```python
def login_required(func):
  def wrapper(*args, **kwargs):
    if "user" in kwargs:
      user_to_check = kwargs["user"]
    elif args:
      user_to_check = args[0]

    if user_to_check == "admin":
      return func(*args, **kwargs)
    else:
      print("Access Denied")
  return wrapper

@login_required
def access_deshboard(user):
  print(f"{user} access")

access_deshboard(user="admin")
```

***Question: Create a decorator that logs function name and parameters. The function should take exactly 2 arguments.***

```python
def log_decorator(func):
  def wrapper(a, b):
    print(f"Calling {func.__name__} with arguments {a}, {b}")
    return func(a, b)
  return wrapper

@log_decorator
def multiply(x , y):
  return x * y

result = multiply(5, 6)
print(result)
```

## Time Decorator.

***Question: Create a decorator that measures execution time of a function with no parameters.***

```python
import time
def time_decorator(func):
  def wrapper():
    start_time = time.time()
    result = func()
    end_time = time.time()
    print(f"Time Taken : {end_time - start_time:.4f} second")
    return result
  return wrapper

@time_decorator
def do_work():
  time.sleep(5)
  print("Work Done")

do_work()
```

## Authorization Decorator.

***Question: Create a decorator that checks if a user is "admin".***

```python
def requires_admin(func):
  def wrapper(user):
    if user != "admin":
      print("Access Denied")
      return
    return func(user)
  return wrapper

@requires_admin
def access_deshboard(user):
  print(f"{user} access the deshboard")

access_deshboard("user")
access_deshboard("admin")
```

## Memoization Decorator

***Question: Create a memoization decorator for a function that takes one integer parameter.***

```python
def decor(func):
  cache = {}
  def wrapper(n):
    if n in cache:
      return cache[n]
    result = func(n)
    cache[n] = result
    return result
  return wrapper

@decor
def factorial(n):
  if n == 0:
    return 1
  return n * factorial(n - 1)
print(factorial(5))
```

## Returning Values.

***Decorators can also return values from the wrapped function:***

```python
def double_result(func):
  def wrapper(*arge, **kwargs):
    result = func(*arge, **kwargs)
    return result * 2
  return wrapper

@ double_result
def get_numbers():
  return 5

print(get_numbers())
```
